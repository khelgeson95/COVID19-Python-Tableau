# 02. Exploring Relationships

## Correlation matrix heatmap - COVID Deaths

```python
# create a correlation matrix using pandas with the covid deaths dataframe
df_covid_deaths.corr()
```

```python
# check columns in covid deaths dataframe
df_covid_deaths.columns
```

```python
df_subset1 = df_covid_deaths[['Year','Month','COVID-19 Deaths', 'Total Deaths', 'Pneumonia Deaths','Pneumonia and COVID-19 Deaths', 'Influenza Deaths', 'Pneumonia, Influenza, or COVID-19 Deaths']]
```

```python
# create a correlation heatmap using matplotlib
f = plt.figure(figsize=(15, 15)) # figure size 
plt.matshow(df_subset1.corr(), fignum=f.number) # type of plot
plt.xticks(range(df_subset1.shape[1]), df_subset1.columns, fontsize=18, rotation=90) # x axis labels
plt.yticks(range(df_subset1.shape[1]), df_subset1.columns, fontsize=18) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=18) # add font size
plt.title('Correlation Matrix', fontsize=26) # add title
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(df_subset1.corr(), annot = True, ax = ax)
```

### Correlations between Year and Month with any of the death total columns are very minimal as the coefficients are very close to 0 here. This means there is not an obvious trend here, so time of year does not appear to have a positive or negative relationship with death totals. There are correlations with coefficients 0.5 - 1.0 for the following relationships: COVID-19 Deaths vs Total Deaths, COVID-19 Deaths vs Pneumonia Deaths, COVID-19 Deaths vs Pneumonia and COVID-19 Deaths, COVID-19 vs Pneumonia, Influenza, or COVID-19 Deaths. This means there is a positive, strong correlation between those variables. There are weak to moderate relationships (coefficients between 0.1 - 0.4) between the following variables: Influenza Deaths vs. COVID-19 Deaths, Influenza Deaths vs. Total Deaths, Influenza Deaths vs Pneumonia Deaths, Influenza Deaths vs Pneumonia, Influenza, or COVID-19 Deaths.

## Scatterplots - COVID Deaths

### There is more variability between these variables. More outliers, more points further from the plotted line.

```python
# Create a scatterplot for the "Influenza Deaths" and "Total Deaths" columns in seaborn
sns.lmplot(x = 'Influenza Deaths', y = 'Total Deaths', data = df_covid_deaths)
```

### Not a strong relationship here.

```python
# Create a scatterplot for the "COVID-19 Deaths" and "Pneumonia and COVID-19 Deaths" columns in seaborn
sns.lmplot(x = 'COVID-19 Deaths', y = 'Pneumonia and COVID-19 Deaths', data = df_covid_deaths)
```

### Very strong relationship here. Not many outliers that stray too far from the line. More COVID-19 Deaths --> More Pneumonia and COVID-19 Deaths (not necessarily a causal relationship, just a strong correlation worth investigating to figure out the exact meaning. It could just mean that if more people are dying from COVID-19 then more people will have died from Pneumonia and COVID-19, but it could also mean that Pneumonia and COVID-19 are closely related illnesses.)

## Categorical Plot - COVID Deaths

```python
# Use a histogram to visualize the distribution of the variables. 
sns.histplot(df_covid_deaths['COVID-19 Deaths'], bins = 100, kde = True)
```

```python
# create 'Low COVID Death Count' label
df_covid_deaths.loc[df_covid_deaths['COVID-19 Deaths'] < 200, 'COVID Death Count Category'] = 'Low COVID Death Count'
```

```python
# create 'Medium COVID Death Count' label
df_covid_deaths.loc[(df_covid_deaths['COVID-19 Deaths'] >= 200) & (df_covid_deaths['COVID-19 Deaths'] < 600), 'COVID Death Count Category'] = 'Medium COVID Death Count'
```

```python
# create 'High COVID Death Count' label
df_covid_deaths.loc[df_covid_deaths['COVID-19 Deaths'] >= 600, 'COVID Death Count Category'] = 'High COVID Death Count'
```

```python
# check frequency of new 'COVID Death Count Category' flag
df_covid_deaths['COVID Death Count Category'].value_counts(dropna = False)
```

```python
df_covid_deaths.head()
```

```python
    #filter out rows with "All Ages" in the "Age Group" column
filtered_df_1 = df_covid_deaths[df_covid_deaths['Age Group'] != 'All Ages']
```

```python
# export dataframe to .csv file for tableau graph
output_file_path = os.path.join(path, '02 Data', 'Prepared Data', 'COVID_Death_Count_Category.csv')
filtered_df_1.to_csv(output_file_path, index=False)
```

```python
# Create a categorical plot in seaborn
sns.set(style="ticks")
g = sns.catplot(x="Age Group", y="COVID-19 Deaths", hue="COVID Death Count Category", data=df_covid_deaths)
```

### The labels overlap but knowing the first one is 'All Ages' followed by the age groups in order, the older age groups more often have High COVID Death Counts.

```python
# Create a categorical plot in seaborn
sns.set(style="ticks")
h = sns.catplot(x="Month", y="COVID-19 Deaths", hue="COVID Death Count Category", data=df_covid_deaths)
```

### Based off of this visualization, there does not appear to be a specific season where there are more COVID-19 Deaths.

```python
# Create a categorical plot in seaborn
sns.set(style="ticks")
h = sns.catplot(x="Year", y="COVID-19 Deaths", hue="COVID Death Count Category", data=df_covid_deaths)
```

### Overall the most COVID-19 deaths occurred 2020 and 2021. We saw less High COVID Death Counts in 2022. 2023 is only partially recorded.

# Categorical Plot - Contributing Conditions

```python
# create 'Low COVID Death Count' label
df_conditions.loc[df_conditions['COVID-19 Deaths'] < 200, 'COVID Death Count Category'] = 'Low COVID Death Count'
```

```python
# create 'Medium COVID Death Count' label
df_conditions.loc[(df_conditions['COVID-19 Deaths'] >= 200) & (df_conditions['COVID-19 Deaths'] < 600), 'COVID Death Count Category'] = 'Medium COVID Death Count'
```

```python
# create 'High COVID Death Count' label
df_conditions.loc[df_conditions['COVID-19 Deaths'] >= 600, 'COVID Death Count Category'] = 'High COVID Death Count'
```

```python
# check frequency of new 'COVID Death Count Category' flag
df_conditions['COVID Death Count Category'].value_counts(dropna = False)
```

```python
df_conditions.head()
```

```python
# export dataframe to .csv file for tableau graph
output_file_path = os.path.join(path, '02 Data', 'Prepared Data', 'Conditions_Death_Count_Category.csv')
df_conditions.to_csv(output_file_path, index=False)
```

```python
# Create a categorical plot in seaborn
sns.set(style="ticks")
k = sns.catplot(x="Condition", y="COVID-19 Deaths", hue="COVID Death Count Category", data=df_conditions, height=6, aspect=2)
k.set_xticklabels(rotation=90)
```

### Conditions contributing the most to COVID-19 death counts from 2020 - 2023 in order from greatest to least are: Influenza and pneumonia, respiratory failure, cardiac arrest, diabetes, and hypertensive diseases.

```python
# Create a categorical plot in seaborn
sns.set(style="ticks")
k = sns.catplot(x="Condition Group", y="COVID-19 Deaths", hue="COVID Death Count Category", data=df_conditions, height=6, aspect=2)
k.set_xticklabels(rotation=90)
```

### Condition groups contributing the most to COVID-19 death counts from 2020 - 2023 in order from greatest to least are: Respiratory Diseases, Circulatory Diseases, Diabetes, Renal failure, and Sepsis.
