# 10. Replace Null Values

```python
# create list of columns to replace null values
columns_to_fill = ['COVID-19 Deaths']
```

```python
# create new dataframe that has all null values replaced by random integer from 1 to 9 inclusive
df_conditions_month_replaced = df_conditions_month_3[columns_to_fill].fillna(np.random.randint(1,10))
```

```python
# add the other columns that were not changed from df_conditions_month_3 dataframe
df_conditions_month_replaced = pd.concat([df_conditions_month_3.drop(columns=columns_to_fill), df_conditions_month_replaced], axis=1)
```

# 11. Check COVID-19 Outliers

```python
# calculating quartiles
Q1 = df_conditions_month_replaced['COVID-19 Deaths'].quantile(0.25)
Q2 = df_conditions_month_replaced['COVID-19 Deaths'].quantile(0.50)
Q3 = df_conditions_month_replaced['COVID-19 Deaths'].quantile(0.75)
```

```python
# show calculated quartiles
print("Q1:", Q1)
print("Q2:", Q2)
print("Q3:", Q3)
```

```python
# calculate standard deviation
std_dev = df_conditions_month_replaced['COVID-19 Deaths'].std()
```

```python
# show standard deviation
std_dev
```

```python
# calculate mean
mean = df_conditions_month_replaced['COVID-19 Deaths'].mean()
```

```python
# show mean
mean
```

```python
# calculate two standard deviations above mean (upper bound)
upper_bound = mean + (2 * std_dev)
```

```python
# show upper bound
upper_bound
```

```python
# calculate two standard deviations below mean (lower bound)
lower_bound = mean - (2 * std_dev)
```

```python
# show lower bound
lower_bound
```

### The lower bound here is negative which is not sensical when talking about population deaths, so the lower bound is effectively 0.

```python
# define outliers
COVID_outliers = df_conditions_month_replaced[(df_conditions_month_replaced['COVID-19 Deaths'] < lower_bound) | (df_conditions_month_replaced['COVID-19 Deaths'] > upper_bound)]
```

```python
# show resulting dataframe containing only COVID-19 Death outliers
COVID_outliers
```

```python
COVID_outliers.info()
```

### There are 5875 entries of total 511980 that are considered outliers for COVID-19 Death Counts. We will leave these outliers alone since this data is sourced and collected reliably and does not indicate any error in the data.
